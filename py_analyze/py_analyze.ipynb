```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Все таблицы csv
df_graduates = pd.read_csv('D:/Mentors/database/Tables/graduates.csv')
df_mentors = pd.read_csv('D:/Mentors/database/Tables/mentors.csv')
df_skills_list = pd.read_csv('D:/Mentors/database/Tables/skills_list.csv')
df_graduate_problems = pd.read_csv('D:/Mentors/database/Tables/graduate_problems.csv')
df_mentor_competencies = pd.read_csv('D:/Mentors/database/Tables/mentor_competencies.csv')
df_pairs = pd.read_csv('D:/Mentors/database/Tables/pairs.csv')
df_meetings = pd.read_csv('D:/Mentors/database/Tables/meetings.csv')
```

```python
#Кол-во строк в таблицах
print("Кол-во выпускников:", len(df_graduates))
print("Наставников:", len(df_mentors))
print("Встреч:", len(df_meetings))
print("Пар:", len(df_pairs))
print("Навыков:", len(df_skills_list))
```

```python
# Портрет участника программы - выпускника

#Гендерное распределение
print("Количественное соотношение мужчины-женщины")
gender_count = df_graduates['gender'].value_counts()
print("Женщины:", gender_count.get('F', 0))
print("Мужчины:", gender_count.get('M', 0))

#Круговая диаграмма - гендерное распределение
plt.figure(figsize=(6, 6))
gender_data = [gender_count.get('F', 0), gender_count.get('M', 0)]
plt.pie(gender_data, labels=['Женщины', 'Мужчины'], autopct='%1.1f%%', startangle=90, colors=['pink', 'blue'])
plt.title('Гендерное распределение выпускников')
plt.tight_layout()
plt.savefig('gender_pie.png')
plt.show()

#Статус занятости в процентах
print("\nСтатус занятости:")
employment_counts = df_graduates['employment_status'].value_counts()
for status, count in employment_counts.items():
    print(f"{status}:{count} ({count/len(df_graduates)*100:.1f}%)")

# Столбчатая диаграмма - статус занятости
plt.figure(figsize=(8, 5))
status_labels = employment_counts.index.tolist()
status_values = employment_counts.values.tolist()
plt.bar(status_labels, status_values, color=['turquoise', 'skyblue', 'palegreen', 'mediumaquamarine'])
plt.xlabel('Статус занятости')
plt.ylabel('Количество выпускников')
plt.title('Статус занятости выпускников')
for i, v in enumerate(status_values):
    plt.text(i, v + 0.1, str(v), ha='center')
plt.tight_layout()
plt.savefig('employment_bar.png')
plt.show()

#Уровень образования в процентах
print("\nУровень образования:")
edu_counts = df_graduates['education_level'].value_counts()
for edu, count in edu_counts.items():
    print(f"{edu}:{count} ({count/len(df_graduates)*100:.1f}%)")

# Круговая диаграмма - уровень обр. в процентах
plt.figure(figsize=(7, 7))
edu_labels = edu_counts.index.tolist()
edu_values = edu_counts.values.tolist()
plt.pie(edu_values, labels=edu_labels, autopct='%1.1f%%', startangle=90, colors=['deepskyblue', 'turquoise', 'skyblue', 'palegreen', 'mediumaquamarine'])
plt.title('Уровень образования выпускников')
plt.tight_layout()
plt.savefig('education_pie.png')
plt.show()
```
